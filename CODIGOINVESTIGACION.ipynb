# DESCRIPTIVOS: dfSecuencia

```python
import pandas as pd

# ============================================================
# RESUMEN (Autoeficacia): T1, T2 y T3 (final) + cambios (deltas)
# - pollCount: 1 = T1 (primera medición), 2 = T2 (segunda), 3 = T3 (final)
# - pollValue: puntaje de autoeficacia en esa medición
# - grupo_intervencion: grupo asignado (para comparar medias y SD)
# Resultado: tabla con Media (SD) por grupo para T1, T2, T3 y deltas
# ============================================================

# 1) Filtrar solo las tres mediciones que nos interesan (T1, T2, T3)
df = dfSecuencia[dfSecuencia["pollCount"].isin([1, 2, 3])].copy()

# 2) Pasar a formato "wide": una fila por persona, una columna por tiempo
se_wide = (
    df.pivot_table(index="userid", columns="pollCount", values="pollValue", aggfunc="mean")
      .rename(columns={1: "SE_T1", 2: "SE_T2", 3: "SE_T3_final"})
)

# 3) Calcular cambios (deltas) entre tiempos
se_wide["dSE_T1_T2"]      = se_wide["SE_T2"]      - se_wide["SE_T1"]
se_wide["dSE_T2_T3_final"] = se_wide["SE_T3_final"] - se_wide["SE_T2"]
se_wide["dSE_T1_T3_final"] = se_wide["SE_T3_final"] - se_wide["SE_T1"]

vars_obj = ["SE_T1", "SE_T2", "SE_T3_final", "dSE_T1_T2", "dSE_T2_T3_final", "dSE_T1_T3_final"]

# 4) Unir el grupo (una fila por persona) y quedarnos con usuarios con grupo definido
df_est = (
    df[["userid", "grupo_intervencion"]].drop_duplicates("userid")
      .merge(se_wide.reset_index(), on="userid", how="left")
      .dropna(subset=["grupo_intervencion"])
)

# 5) Función para formatear cada variable como "Media (SD)"
def mean_sd(s: pd.Series) -> str:
    s = s.dropna()
    return "—" if s.empty else f"{s.mean():.2f} ({s.std(ddof=1):.2f})"

# 6) Tabla final: filas = variables, columnas = grupos
tabla = df_est.groupby("grupo_intervencion")[vars_obj].agg(mean_sd).T

# 7) Etiquetas para reporte
tabla.index = tabla.index.map({
    "SE_T1": "SEₜ₁ (T1)",
    "SE_T2": "SEₜ₂ (T2)",
    "SE_T3_final": "SEₜ₃ (Final)",
    "dSE_T1_T2": "∆SEₜ₁,₂",
    "dSE_T2_T3_final": "∆SEₜ₂,₃",
    "dSE_T1_T3_final": "∆SEₜ₁,₃",
})

print("\nP2. Variables de Autoeficacia (T1, T2, T3-final)")
print(tabla)
```

# DESCRIPTIVOS = dfUsuario

```python
import pandas as pd
import numpy as np

# ============================================================
# P3.1 y P3.2 (dfUsuario) — resumen por grupo
# Objetivo:
#   - Construir variables derivadas a partir de:
#       % de tópicos (arriba/abajo)  x  total de casos
#   - Resumir por grupo_intervencion en una tabla: Media o Media (SD)
#
# Nota:
#   A) FUNCIONES DE ANÁLISIS (transforman datos, afectan resultados)
#   B) FUNCIONES ESTÉTICAS (solo formatean texto para reporte; NO cambian resultados)
# ============================================================

df = dfUsuario.copy()

# ============================================================
# A) FUNCIONES DE ANÁLISIS
# ============================================================

def pct_times_total(d: pd.DataFrame, pct_col: str, total_col: str = "casos") -> pd.Series:
    """
    Convierte porcentaje/proporción a 'conteo esperado' multiplicando por total.
    Soporta dos escalas en pct_col:
      - 0..1  (proporción)  -> usa pct tal cual
      - 0..100 (porcentaje) -> divide por 100
    Detecta escala con una heurística: mediana > 1 => asume 0..100.
    """
    pct = pd.to_numeric(d.get(pct_col), errors="coerce")
    tot = pd.to_numeric(d.get(total_col), errors="coerce")

    if pct is None or tot is None:
        return pd.Series(np.nan, index=d.index)

    scale_0_100 = pct.dropna().median() > 1 if pct.notna().any() else False
    prop = pct / 100 if scale_0_100 else pct
    return prop * tot


# ============================================================
# B) FUNCIONES ESTÉTICAS
# ============================================================

def _to_numeric_clean(s: pd.Series) -> pd.Series:
    """Convierte a numérico y elimina NA. Utilidad interna para formateo."""
    return pd.to_numeric(s, errors="coerce").dropna()

def fmt_mean(s: pd.Series) -> str:
    """Devuelve 'Media' con 2 decimales o '—' si no hay datos."""
    s = _to_numeric_clean(s)
    return "—" if s.empty else f"{s.mean():.2f}"

def fmt_mean_sd(s: pd.Series) -> str:
    """Devuelve 'Media (SD)' con 2 decimales o '—' si no hay datos."""
    s = _to_numeric_clean(s)
    if s.empty:
        return "—"
    sd = s.std(ddof=1)
    sd = 0.0 if pd.isna(sd) else sd  # si hay un solo dato, SD queda NaN -> lo mostramos como 0.00
    return f"{s.mean():.2f} ({sd:.2f})"


# ============================================================
# 1) VARIABLES DERIVADAS (ANÁLISIS)
# ============================================================
# Subtópicos esperados = proporción/porcentaje de tópicos * total de casos
df["Subtopicos_UW_prom"] = pct_times_total(df, pct_col="porce_topicos_arriba", total_col="casos")
df["Subtopicos_DW_prom"] = pct_times_total(df, pct_col="porce_topicos_abajo", total_col="casos")


# ============================================================
# 2) ESPECIFICACIÓN DE TABLA 
# ============================================================
# Cada fila: (Etiqueta que aparece en la tabla, columna en df, tipo de resumen)
# - "mean"    => solo media
# - "mean_sd" => media y desviación estándar
rows = [
    ("Subtópicos Upward (UW; prom.)",        "Subtopicos_UW_prom",               "mean"),
    ("Subtópicos Downward (DW; prom.)",      "Subtopicos_DW_prom",               "mean"),

    ("Tuw (casos arriba)",                   "casos_comparacion_arriba",         "mean_sd"),
    ("Tdw (casos abajo)",                    "casos_comparacion_abajo",          "mean_sd"),

    ("PropTup",                              "porce_topicos_arriba",             "mean_sd"),
    ("PropTdown",                            "porce_topicos_abajo",              "mean_sd"),

    ("PrefTcomp",                            "preferencia_casos_comparacion",    "mean_sd"),
    ("PrefTup",                              "preferencia_casos_uw",             "mean_sd"),

    ("Steps",                                "pasos_reales_ejercicio",           "mean_sd"),
    ("Att1c",                                "steps_primera",                    "mean_sd"),

    ("PerfAtt",                              "performance",                      "mean_sd"),
    ("PerfAttup",                            "performance_arriba",               "mean_sd"),
    ("PerfAttdw",                            "performance_abajo",                "mean_sd"),
]


# ============================================================
# 3) CONSTRUIR TABLA RESUMEN POR GRUPO
# ============================================================
# Orden explícito de columnas
orden_grupos = ["OLM", "OSLM", "OSLM+VP"]

# Tabla vacía (rellena con "—" por defecto)
tabla = pd.DataFrame(index=[label for (label, _, _) in rows], columns=orden_grupos).fillna("—")

# Llenado: por cada grupo, calcular el resumen indicado por cada fila
for g, sub in df.groupby("grupo_intervencion"):
    if g not in tabla.columns:
        continue

    for label, col, how in rows:
        if col not in sub.columns:
            continue

        tabla.loc[label, g] = fmt_mean(sub[col]) if how == "mean" else fmt_mean_sd(sub[col])


print("\nP3.1 y P3.2 (dfUsuario) — resumen por grupo (Steps = pasos_reales_ejercicio)")
print(tabla)
```

# MIXEDLM - GENERAL

```python
import pandas as pd
import statsmodels.formula.api as smf

formula = (
    "pollValue_next ~ "
    "pollValue + "
    "steps_primera + "
    "last_progressMe + "
    "selectionRating_result + "
    "comparacionOSLM + "
    "comparacionOSLMVP + "
    "proporcion_pasos_primera_vez"
)


# ------------------------------------------------------------
# Variables que deben ser numéricas para el modelo
# ------------------------------------------------------------
columnas_numericas = [
    "pollValue_next",
    "pollValue",
    "steps_primera",
    "last_progressMe",
    "selectionRating_result",
    "proporcion_pasos_primera_vez",
]


# ------------------------------------------------------------
# Preparación del dataset para el modelo (ANÁLISIS)
# ------------------------------------------------------------
df_modelo = (
    dfSecuencia

    # A) Eliminar filas sin outcome o sin rating (no se puede estimar el modelo)
    .dropna(subset=["pollValue_next", "selectionRating_result"])

    # B) Crear variables dummy de comparación entre grupos
    #    comparacionOSLM   = 1 si NO es OLM (OSLM u OSLM+VP)
    #    comparacionOSLMVP = 1 solo si es OSLM+VP
    .assign(
        comparacionOSLM=lambda d: (d["grupo_intervencion"] != "OLM").astype(int),
        comparacionOSLMVP=lambda d: (d["grupo_intervencion"] == "OSLM+VP").astype(int),
    )

    # C) Asegurar que las variables relevantes sean numéricas
    .assign(**{
        col: lambda d, col=col: pd.to_numeric(d[col], errors="coerce")
        for col in columnas_numericas
    })

    # D) Seleccionar solo columnas necesarias para el modelo
    .loc[:, (
        ["userid"]
        + columnas_numericas
        + ["comparacionOSLM", "comparacionOSLMVP"]
    )]

    # E) Eliminar filas con cualquier NA restante
    .dropna()

    # F) Asegurar que userid sea categórica (requerido por MixedLM)
    .assign(userid=lambda d: d["userid"].astype("category"))

    # G) Resetear índice para limpieza final
    .reset_index(drop=True)
)


# ------------------------------------------------------------
# Ajuste del modelo mixto
# ------------------------------------------------------------
# - groups = userid → intercepto aleatorio por persona
modelo_mixto = smf.mixedlm(
    formula=formula,
    data=df_modelo,
    groups=df_modelo["userid"]
)

resultado = modelo_mixto.fit()


# ------------------------------------------------------------
# Resumen del modelo
# ------------------------------------------------------------
print(resultado.summary())
```

# Pregunta 3.1

```python
import pandas as pd
import numpy as np
from scipy import stats

# ------------------------------------------------------------
# 0) Parámetros del análisis (grupos y variables)
# ------------------------------------------------------------
GRUPOS_3 = ["OLM", "OSLM", "OSLM+VP"]      # para ANOVA
GRUPOS_2 = ["OSLM", "OSLM+VP"]            # para comparaciones 2 grupos

# Variables P3.1: etiquetas nombre de columna en dfUsuario
VARS_P31 = {
    "PropTup":   "porce_topicos_arriba",
    "PropTdown": "porce_topicos_abajo",
    "PrefTcomp": "preferencia_casos_comparacion",
    "PrefTup":   "preferencia_casos_uw",
}

# Variables de progreso
PROG_IND = "progreso_inicial"        # ANOVA 3 grupos
PROG_GRP = "progreso_inicial_grupo"  # MW 2 grupos


# ------------------------------------------------------------
# 1) Datos (filtrados por grupos)
# ------------------------------------------------------------
df_all = dfUsuario.loc[dfUsuario["grupo_intervencion"].isin(GRUPOS_3)].copy()
df_2g  = df_all.loc[df_all["grupo_intervencion"].isin(GRUPOS_2)].copy()


# ============================================================
# A) FUNCIONES ESTADÍSTICAS
# ============================================================

def to_num(s: pd.Series) -> pd.Series:
    """Convierte a numérico y elimina NA (base para todos los tests)."""
    return pd.to_numeric(s, errors="coerce").dropna()

def mann_whitney_u(a: pd.Series, b: pd.Series):
    """
    Test Mann–Whitney U (2 muestras independientes).
    Devuelve: (U, p). Si no hay datos suficientes: (nan, nan)
    """
    a, b = to_num(a), to_num(b)
    if a.empty or b.empty:
        return (np.nan, np.nan)
    try:
        U, p = stats.mannwhitneyu(a, b, alternative="two-sided", method="auto")
    except TypeError:
        # compatibilidad con versiones antiguas de SciPy
        U, p = stats.mannwhitneyu(a, b, alternative="two-sided")
    return (U, p)

def one_way_anova(*groups: pd.Series):
    """
    ANOVA de un factor (3+ grupos).
    Devuelve: (F, p, ns_por_grupo)
    """
    gs = [to_num(g).values for g in groups]
    ns = [len(g) for g in gs]
    if any(n == 0 for n in ns):
        return (np.nan, np.nan, ns)
    F, p = stats.f_oneway(*gs)
    return (F, p, ns)

def normality_by_group(df: pd.DataFrame, var: str, groups: list[str]) -> pd.DataFrame:
    """
    Chequeo de normalidad por grupo para una variable:
      - Shapiro-Wilk (n >= 3)  [se capea a 5000 para evitar problemas]
      - D’Agostino K^2 (n >= 8)
      - Correlación QQ (n >= 3) como indicador gráfico-numérico
    Retorna DataFrame con métricas por grupo.
    """
    out = []

    for g in groups:
        s = to_num(df.loc[df["grupo_intervencion"] == g, var])
        n = len(s)

        # Shapiro (>=3). Cap 5000 como práctica común.
        if n >= 3:
            s_sh = s.sample(min(n, 5000), random_state=123) if n > 5000 else s
            W, pW = stats.shapiro(s_sh)
        else:
            W, pW = np.nan, np.nan

        # D’Agostino (>=8)
        K2, pK2 = stats.normaltest(s) if n >= 8 else (np.nan, np.nan)

        # QQ corr (>=3)
        if n >= 3:
            osm, osr = stats.probplot(s, dist="norm")[0]
            rqq = np.corrcoef(osm, osr)[0, 1]
        else:
            rqq = np.nan

        out.append({
            "Variable": var,
            "Grupo": g,
            "n": n,
            "Shapiro_W": W,
            "Shapiro_p": pW,
            "Normaltest_K2": K2,
            "Normaltest_p": pK2,
            "QQ_corr_r": rqq,
        })

    return pd.DataFrame(out)


# ============================================================
# B) FUNCIONES ESTÉTICAS
# ============================================================

def mean_sd_str(s: pd.Series) -> str:
    """Devuelve 'Media (SD)' con 2 decimales o '—' si no hay datos."""
    s = to_num(s)
    if s.empty:
        return "—"
    sd = s.std(ddof=1)
    sd = 0.0 if pd.isna(sd) else sd
    return f"{s.mean():.2f} ({sd:.2f})"

def sig_star(p) -> str:
    """Marca significancia con * si p < .05 (solo decorativo)."""
    return "*" if (pd.notna(p) and p < 0.05) else ""


# ============================================================
# 2) Table 4 (OSLM vs OSLM+VP) — SOLO P3.1 (DESCRIPTIVOS)
# ============================================================
# Esto es descriptivo (reporte), pero usa mean_sd_str (estético) sobre datos reales.
tabla4 = pd.DataFrame({
    g: {
        etiqueta: mean_sd_str(df_2g.loc[df_2g["grupo_intervencion"] == g, col])
        for etiqueta, col in VARS_P31.items()
    }
    for g in GRUPOS_2
})


# ============================================================
# 3) Tests principales (ESTADÍSTICOS)
# ============================================================

# 3.1) Mann–Whitney U para variables P3.1 (OSLM vs OSLM+VP)
mw_tests = {
    etiqueta: mann_whitney_u(
        df_2g.loc[df_2g["grupo_intervencion"] == "OSLM", col],
        df_2g.loc[df_2g["grupo_intervencion"] == "OSLM+VP", col]
    )
    for etiqueta, col in VARS_P31.items()
}

# 3.2) ANOVA para PROG_IND (3 grupos)
F_progInd, p_progInd, ns_progInd = one_way_anova(
    df_all.loc[df_all["grupo_intervencion"] == "OLM", PROG_IND],
    df_all.loc[df_all["grupo_intervencion"] == "OSLM", PROG_IND],
    df_all.loc[df_all["grupo_intervencion"] == "OSLM+VP", PROG_IND],
)

# 3.3) Mann–Whitney U para PROG_GRP (OSLM vs OSLM+VP)
U_progGrp, p_progGrp = mann_whitney_u(
    df_all.loc[df_all["grupo_intervencion"] == "OSLM", PROG_GRP],
    df_all.loc[df_all["grupo_intervencion"] == "OSLM+VP", PROG_GRP],
)


# ============================================================
# 4) Chequeo de normalidad (SUPUESTOS) — ESTADÍSTICO
# ============================================================
norm_anova = normality_by_group(df_all, PROG_IND, GRUPOS_3)

norm_2g = pd.concat(
    [normality_by_group(df_all, v, GRUPOS_2) for v in list(VARS_P31.values()) + [PROG_GRP]],
    ignore_index=True
)


# ============================================================
# 5) Salida (ESTÉTICA / REPORTE)
# ============================================================

# Tamaños muestrales (N únicos por usuario) para el reporte
n_oslm = df_2g.loc[df_2g["grupo_intervencion"] == "OSLM", "userid"].nunique()
n_vp   = df_2g.loc[df_2g["grupo_intervencion"] == "OSLM+VP", "userid"].nunique()

print("\nTable 4. Proporciones y preferencias en selección de subtópicos con comparación social")
print(f"OSLM (N={n_oslm}) vs OSLM+VP (N={n_vp})\n")
print(tabla4)

# Reporte MW para variables P3.1
for etiqueta, (U, p) in mw_tests.items():
    print(f"Mann–Whitney U {etiqueta} (OSLM vs OSLM+VP): U = {U:.1f}, p = {p:.3f} {sig_star(p)}")

# Reporte ANOVA (armar df1, df2 con N total y k grupos)
k = 3
N_anova = sum(ns_progInd) if all(isinstance(x, (int, np.integer)) for x in ns_progInd) else np.nan
df1 = k - 1
df2 = (N_anova - k) if pd.notna(N_anova) else np.nan

print(f"\nANOVA ProgInd (OLM, OSLM, OSLM+VP): F({df1}, {df2}) = {F_progInd:.3f}, p = {p_progInd:.3f}")
print(f"Mann–Whitney U ProgGrp (OSLM vs OSLM+VP): U = {U_progGrp:.1f}, p = {p_progGrp:.3f}")

print("\n========================")
print("Chequeo de normalidad (supuestos)")
print("========================")

print("\n-- Para ANOVA (ProgInd) por grupo --")
print(norm_anova[["Variable","Grupo","n","Shapiro_W","Shapiro_p","Normaltest_K2","Normaltest_p","QQ_corr_r"]])

print("\n-- Para comparación OSLM vs OSLM+VP (variables clave) --")
print(norm_2g[["Variable","Grupo","n","Shapiro_W","Shapiro_p","Normaltest_K2","Normaltest_p","QQ_corr_r"]])
```

# PREGUNTA 3.2

```python
import pandas as pd
import numpy as np
from scipy import stats

# ------------------------------------------------------------
# 0) Parámetros del análisis (grupos y columnas)
# ------------------------------------------------------------
GRUPOS_3 = ["OLM", "OSLM", "OSLM+VP"]      # análisis global (3 grupos)
GRUPOS_2 = ["OSLM", "OSLM+VP"]            # comparaciones específicas (2 grupos)

# Etiquetas “cortas” -> columnas reales en dfUsuario
COLS = {
    "PerfAtt":   "performance",
    "PerfAttup": "performance_arriba",
    "PerfAttdw": "performance_abajo",
}


# ------------------------------------------------------------
# 1) Datos (filtrados) + chequeo de columnas
# ------------------------------------------------------------
df = dfUsuario.loc[dfUsuario["grupo_intervencion"].isin(GRUPOS_3)].copy()

faltantes = [c for c in COLS.values() if c not in df.columns]
if faltantes:
    raise KeyError(f"Faltan columnas en dfUsuario: {faltantes}")


# ============================================================
# A) FUNCIONES ESTADÍSTICAS
# ============================================================

def to_num(s: pd.Series) -> pd.Series:
    """Convierte a numérico y elimina NA; base para tests y descriptivos."""
    return pd.to_numeric(s, errors="coerce").dropna()

def mann_whitney_u(a: pd.Series, b: pd.Series):
    """
    Mann–Whitney U (2 grupos).
    Retorna: (U, p). Si faltan datos: (nan, nan)
    """
    a, b = to_num(a), to_num(b)
    if a.empty or b.empty:
        return (np.nan, np.nan)
    try:
        U, p = stats.mannwhitneyu(a, b, alternative="two-sided", method="auto")[:2]
    except TypeError:
        U, p = stats.mannwhitneyu(a, b, alternative="two-sided")[:2]
    return (U, p)

def kruskal_wallis_3grupos(df_: pd.DataFrame, col: str, grupos: list[str]):
    """
    Kruskal–Wallis (3+ grupos), alternativa no paramétrica a ANOVA.
    Retorna: (H, p). Si falta algún grupo: (nan, nan)
    """
    arrays = [
        to_num(df_.loc[df_["grupo_intervencion"] == g, col]).values
        for g in grupos
    ]
    if any(len(a) == 0 for a in arrays):
        return (np.nan, np.nan)
    H, p = stats.kruskal(*arrays)
    return (H, p)

def shapiro_wilk(s: pd.Series):
    """
    Shapiro–Wilk por grupo:
      - requiere n >= 3
      - si n > 5000, se samplea 5000 (para estabilidad/tiempo)
    Retorna: (n_usado, W, p)
    """
    s = to_num(s)
    n = len(s)
    if n < 3:
        return (n, np.nan, np.nan)
    if n > 5000:
        s = s.sample(5000, random_state=123)
    W, p = stats.shapiro(s)
    return (len(s), W, p)


# ============================================================
# B) FUNCIONES ESTÉTICAS (solo presentación / no cambian análisis)
# ============================================================

def mean_sd_str(s: pd.Series) -> str:
    """Devuelve 'Media (SD)' con 2 decimales o '—' si no hay datos."""
    s = to_num(s)
    if s.empty:
        return "—"
    sd = s.std(ddof=1)
    sd = 0.0 if pd.isna(sd) else sd
    return f"{s.mean():.2f} ({sd:.2f})"


# ============================================================
# 2) Tabla descriptiva (ESTÉTICA + datos reales)
# ============================================================
# Estructura:
#   - Filas: PerfAtt, PerfAttup, PerfAttdw
#   - Columnas: OLM, OSLM, OSLM+VP (pero up/dw solo para 2 grupos)
tabla = pd.DataFrame(index=COLS.keys(), columns=GRUPOS_3).fillna("—")

# 2.1) PerfAtt (global) para 3 grupos
for g in GRUPOS_3:
    tabla.loc["PerfAtt", g] = mean_sd_str(
        df.loc[df["grupo_intervencion"] == g, COLS["PerfAtt"]]
    )

# 2.2) PerfAttup y PerfAttdw (solo OSLM vs OSLM+VP)
for g in GRUPOS_2:
    for key in ["PerfAttup", "PerfAttdw"]:
        tabla.loc[key, g] = mean_sd_str(
            df.loc[df["grupo_intervencion"] == g, COLS[key]]
        )


# ============================================================
# 3) Contrastes (ESTADÍSTICOS)
# ============================================================

# 3.1) Kruskal–Wallis para PerfAtt (3 grupos)
H, pH = kruskal_wallis_3grupos(df, COLS["PerfAtt"], GRUPOS_3)

# 3.2) Mann–Whitney U para PerfAttup (2 grupos)
U_up, p_up = mann_whitney_u(
    df.loc[df["grupo_intervencion"] == "OSLM",    COLS["PerfAttup"]],
    df.loc[df["grupo_intervencion"] == "OSLM+VP", COLS["PerfAttup"]],
)

# 3.3) Mann–Whitney U para PerfAttdw (2 grupos)
U_dw, p_dw = mann_whitney_u(
    df.loc[df["grupo_intervencion"] == "OSLM",    COLS["PerfAttdw"]],
    df.loc[df["grupo_intervencion"] == "OSLM+VP", COLS["PerfAttdw"]],
)


# ============================================================
# 4) Chequeo de normalidad (SUPUESTOS) — ESTADÍSTICO
# ============================================================
# Nota: Aunque usamos tests no paramétricos (Kruskal/MW),
# igual es útil reportar normalidad como diagnóstico descriptivo.
norm_rows = []

# 4.1) Normalidad para PerfAtt (3 grupos)
for g in GRUPOS_3:
    n, W, pW = shapiro_wilk(df.loc[df["grupo_intervencion"] == g, COLS["PerfAtt"]])
    norm_rows.append({"Variable": "PerfAtt", "Grupo": g, "n": n, "Shapiro_W": W, "Shapiro_p": pW})

# 4.2) Normalidad para PerfAttup y PerfAttdw (2 grupos)
for g in GRUPOS_2:
    for key in ["PerfAttup", "PerfAttdw"]:
        n, W, pW = shapiro_wilk(df.loc[df["grupo_intervencion"] == g, COLS[key]])
        norm_rows.append({"Variable": key, "Grupo": g, "n": n, "Shapiro_W": W, "Shapiro_p": pW})

norm_df = pd.DataFrame(norm_rows)


# ============================================================
# 5) Salida (ESTÉTICA / REPORTE)
# ============================================================

# N únicos por usuario (para reportar tamaños muestrales)
Ns = {g: df.loc[df["grupo_intervencion"] == g, "userid"].nunique() for g in GRUPOS_3}

print("\nTabla. Performance en actividad (global y por dirección de comparación)")
print(f"OLM (N={Ns['OLM']}) | OSLM (N={Ns['OSLM']}) | OSLM+VP (N={Ns['OSLM+VP']})\n")
print(tabla)

print("\nContrastes")
print(f"Kruskal–Wallis PerfAtt (OLM, OSLM, OSLM+VP): H = {H:.2f}, p = {pH:.3f}")
print(f"Mann–Whitney U PerfAttup (OSLM vs OSLM+VP): U = {U_up:.1f}, p = {p_up:.3f}")
print(f"Mann–Whitney U PerfAttdw (OSLM vs OSLM+VP): U = {U_dw:.1f}, p = {p_dw:.3f}")

print("\nChequeo de normalidad (Shapiro–Wilk)")
print(norm_df)
```

```python
import pandas as pd
import numpy as np

# =========================
# Construcción dfCongreso
# =========================

# Base: 1 fila por usuario con su grupo
dfCongreso = dfNivelEjercicio[["userid", "grupo_intervencion"]].drop_duplicates(subset="userid").copy()

# pollValue a numérico (para autoeficacia)
dfNivelEjercicio["pollValue_num"] = pd.to_numeric(dfNivelEjercicio["pollValue"], errors="coerce")

# -------------------------
# Agregaciones por usuario
# -------------------------

# Número total de pasos (eventos tryStep) por usuario
suma_pasos = dfNivelEjercicio.groupby("userid")["num_steps"].sum(min_count=1)
dfCongreso["num_steps"] = dfCongreso["userid"].map(suma_pasos)

# Número de pasos a la primera por usuario
suma_pasos_primera = dfNivelEjercicio.groupby("userid")["steps_primera"].sum(min_count=1)
dfCongreso["steps_primera"] = dfCongreso["userid"].map(suma_pasos_primera)

# Pasos reales del ejercicio por usuario
# min_count=1 evita que "todo NaN" se transforme silenciosamente en 0
pasos_reales = dfNivelEjercicio.groupby("userid")["pasos_reales_ejercicio"].sum(min_count=1)
dfCongreso["pasos_reales_ejercicio"] = dfCongreso["userid"].map(pasos_reales)

# Número de sesiones distintas
numero_sesiones_diferentes = dfNivelEjercicio.groupby("userid")["session"].nunique()
dfCongreso["numero_sesiones_diferentes"] = dfCongreso["userid"].map(numero_sesiones_diferentes)

# Tasa de éxito (por eventos tryStep)
dfCongreso["Tasa_exito"] = (
    pd.to_numeric(dfCongreso["steps_primera"], errors="coerce") /
    pd.to_numeric(dfCongreso["num_steps"], errors="coerce")
)

# Autoeficacia general (promedio de pollValue numérico)
autoeficacia_inicio = (
    dfNivelEjercicio[dfNivelEjercicio["pollValue_num"].notna()]
    .groupby("userid")["pollValue_num"]
    .mean()
)
dfCongreso["autoeficacia_general"] = dfCongreso["userid"].map(autoeficacia_inicio)

# Progreso inicial (promedios)
progreso_inicial = dfNivelEjercicio.groupby("userid")["last_progressMe"].mean()
dfCongreso["progreso_inicial"] = dfCongreso["userid"].map(progreso_inicial)

progreso_grupo = dfNivelEjercicio.groupby("userid")["last_progressGroup"].mean()
dfCongreso["progreso_inicial_grupo"] = dfCongreso["userid"].map(progreso_grupo)

# Número total de hints
numero_hints = dfNivelEjercicio.groupby("userid")["total_hints"].sum(min_count=1)
dfCongreso["numero_hints"] = dfCongreso["userid"].map(numero_hints)

# Tasa de hints (por eventos tryStep)
dfCongreso["tasa_hints"] = (
    pd.to_numeric(dfCongreso["numero_hints"], errors="coerce") /
    pd.to_numeric(dfCongreso["num_steps"], errors="coerce")
)

# Promedio rating
selectionRating_result = dfNivelEjercicio.groupby("userid")["selectionRating_result"].mean()
dfCongreso["selectionRating_result"] = dfCongreso["userid"].map(selectionRating_result)

# Tiempo total (si no hay tiempo, 0)
suma_tiempo = df_filtrado_tiempo.groupby("userid")["duration_winsor"].sum(min_count=1)
dfCongreso["tiempo_total_segundos_winsorizing"] = dfCongreso["userid"].map(suma_tiempo).fillna(0)

# Primer autoeficacia (primer poll por (userid, session, topiclabel) -> promedio por usuario)
df_ordenado = dfNivelEjercicio.sort_values(["userid", "session", "topiclabel", "start_ts"])
df_primer_poll = df_ordenado.groupby(["userid", "session", "topiclabel"], as_index=False).first()
df_primer_poll = df_primer_poll[["userid", "pollValue_num"]].rename(columns={"pollValue_num": "primer_pollValue"})
primer_pollValue = df_primer_poll.groupby("userid")["primer_pollValue"].mean()
dfCongreso["primer_pollValue"] = dfCongreso["userid"].map(primer_pollValue)

# =========================================
# Proporción pasos a la primera (segura)
# =========================================

dfCongreso["pasos_reales_ejercicio"] = pd.to_numeric(dfCongreso["pasos_reales_ejercicio"], errors="coerce")
dfCongreso["steps_primera"] = pd.to_numeric(dfCongreso["steps_primera"], errors="coerce")

# Recorte: steps_primera no puede exceder pasos_reales_ejercicio
dfCongreso["steps_primera"] = np.minimum(
    dfCongreso["steps_primera"],
    dfCongreso["pasos_reales_ejercicio"],
)

den = dfCongreso["pasos_reales_ejercicio"]
num = dfCongreso["steps_primera"]
num_clip = np.minimum(num, den)

dfCongreso["proporcion_pasos_primera_vez"] = 0.0
mask_den = den > 0
dfCongreso.loc[mask_den, "proporcion_pasos_primera_vez"] = num_clip[mask_den] / den[mask_den]

print("max proporción:", dfCongreso["proporcion_pasos_primera_vez"].max())

# =========================
# Casos totales (bloques)
# =========================
casos = dfNivelEjercicio.groupby("userid").size()
dfCongreso["casos"] = dfCongreso["userid"].map(casos)

# =========================================
# Métricas de comparación arriba/abajo
# =========================================
umbral = 0.01

# Pasos arriba (conteo de num_steps en casos arriba)
pasos_comparacion_arriba = dfNivelEjercicio[
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) > umbral)
].groupby("userid")["num_steps"].sum(min_count=1)
dfCongreso["pasos_comparacion_arriba"] = dfCongreso["userid"].map(pasos_comparacion_arriba).fillna(0)

# Pasos abajo
pasos_comparacion_abajo = dfNivelEjercicio[
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) < -umbral)
].groupby("userid")["num_steps"].sum(min_count=1)
dfCongreso["pasos_comparacion_abajo"] = dfCongreso["userid"].map(pasos_comparacion_abajo).fillna(0)

# Casos arriba (número de bloques/casos)
casos_comparacion_arriba = dfNivelEjercicio[
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) > umbral)
].groupby("userid").size()
dfCongreso["casos_comparacion_arriba"] = dfCongreso["userid"].map(casos_comparacion_arriba).fillna(0)

# Casos abajo
casos_comparacion_abajo = dfNivelEjercicio[
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) < -umbral)
].groupby("userid").size()
dfCongreso["casos_comparacion_abajo"] = dfCongreso["userid"].map(casos_comparacion_abajo).fillna(0)

# Proporción pasos primera vez arriba
prop_primera_arriba = dfNivelEjercicio[
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) > umbral)
].groupby("userid")["proporcion_pasos_primera_vez"].sum(min_count=1)
dfCongreso["proporcion_pasos_primera_vez_arriba"] = dfCongreso["userid"].map(prop_primera_arriba).fillna(0)

# Proporción pasos primera vez abajo
prop_primera_abajo = dfNivelEjercicio[
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) < -umbral)
].groupby("userid")["proporcion_pasos_primera_vez"].sum(min_count=1)
dfCongreso["proporcion_pasos_primera_vez_abajo"] = dfCongreso["userid"].map(prop_primera_abajo).fillna(0)

# =========================================
# PERFORMANCE (FIX: segura, sin perder N)
# =========================================

den = pd.to_numeric(dfCongreso["pasos_reales_ejercicio"], errors="coerce")
num = pd.to_numeric(dfCongreso["steps_primera"], errors="coerce")

dfCongreso["performance"] = 0.0
mask = den > 0
dfCongreso.loc[mask, "performance"] = num[mask] / den[mask]

# =========================================
# PERFORMANCE ARRIBA / ABAJO por usuario
# (si den=0 en esos subconjuntos, queda 0)
# =========================================

# ARRIBA
mask_arriba = (
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) > umbral)
)

agg_arriba = (
    dfNivelEjercicio[mask_arriba]
    .groupby("userid")
    .agg(
        steps_primera_sum=("steps_primera", "sum"),
        pasos_reales_sum=("pasos_reales_ejercicio", "sum"),
    )
)

agg_arriba["performance_arriba"] = (
    agg_arriba["steps_primera_sum"] /
    agg_arriba["pasos_reales_sum"].replace(0, np.nan)
)

dfCongreso["performance_arriba"] = dfCongreso["userid"].map(agg_arriba["performance_arriba"]).fillna(0)

# ABAJO
mask_abajo = (
    (dfNivelEjercicio["last_progressGroup"] != -1) &
    ((dfNivelEjercicio["last_progressMe"] - dfNivelEjercicio["last_progressGroup"]) < -umbral)
)

agg_abajo = (
    dfNivelEjercicio[mask_abajo]
    .groupby("userid")
    .agg(
        steps_primera_sum=("steps_primera", "sum"),
        pasos_reales_sum=("pasos_reales_ejercicio", "sum"),
    )
)

agg_abajo["performance_abajo"] = (
    agg_abajo["steps_primera_sum"] /
    agg_abajo["pasos_reales_sum"].replace(0, np.nan)
)

dfCongreso["performance_abajo"] = dfCongreso["userid"].map(agg_abajo["performance_abajo"]).fillna(0)

# =========================================
# Tiempo arriba / abajo (como lo tenías)
# =========================================

tiempo_casos_comparacion_arriba = df_filtrado_tiempo[
    (df_filtrado_tiempo["progressGroup"] != -1) &
    ((df_filtrado_tiempo["progressMe"] - df_filtrado_tiempo["progressGroup"]) > umbral)
].groupby("userid")["duration_winsor"].sum(min_count=1)
dfCongreso["tiempo_casos_comparacion_arriba"] = dfCongreso["userid"].map(tiempo_casos_comparacion_arriba).fillna(0)

tiempo_casos_comparacion_abajo = df_filtrado_tiempo[
    (df_filtrado_tiempo["progressGroup"] != -1) &
    ((df_filtrado_tiempo["progressMe"] - df_filtrado_tiempo["progressGroup"]) < -umbral)
].groupby("userid")["duration_winsor"].sum(min_count=1)
dfCongreso["tiempo_casos_comparacion_abajo"] = dfCongreso["userid"].map(tiempo_casos_comparacion_abajo).fillna(0)
```

```python
import pandas as pd
import numpy as np
import json

# --- Copiar y ordenar DataFrame ---
perfil = perfil_expanded.copy()
perfil["timestamp"] = pd.to_datetime(perfil["timestamp"])
perfil = perfil.sort_values(["userid", "timestamp"])

# --- Merge grupo de intervención inicial ---
def parse_tags(tag_str):
    if pd.isna(tag_str):
        return set()
    items = tag_str.strip("{}").split(",")
    items = [item.strip().lower() for item in items if item.strip() != ""]
    items = ["oslm" if item == "olsm" else item for item in items]  # corregir typo
    return set(items)

def categorize(tags_set):
    has_oslm = "oslm" in tags_set
    has_motiv = "motiv-msg" in tags_set
    if has_oslm and has_motiv:
        return "OSLM+VP"
    elif has_oslm:
        return "OSLM"
    else:
        return "OLM"

grupo_por_estudiante = (
    perfil.groupby("userid", sort=False)["tags"]
    .apply(lambda tags: categorize(set().union(*tags.apply(parse_tags))))
    .reset_index(name="grupo_intervencion")
)

perfil = perfil.merge(grupo_por_estudiante, on="userid", how="left")

# --- Nueva columna: direccionComparacion ---
def calcular_direccion(row):
    me = row.get("ext_progressMe", np.nan)
    grp = row.get("ext_progressGroup", np.nan)

    # si falta cualquiera de los dos, no inventar dirección
    if pd.isna(me) or pd.isna(grp) or grp == -1:
        return np.nan

    if me < grp:
        return "hacia arriba"
    if me > grp:
        return "hacia abajo"
    return "igual"

perfil["direccionComparacion"] = perfil.apply(calcular_direccion, axis=1)

# Propagar grupo_intervencion y direccionComparacion
# - grupo_intervencion es por usuario (robusto)
perfil["grupo_intervencion"] = (
    perfil.groupby("userid", sort=False)["grupo_intervencion"].ffill().bfill()
)

# - direccionComparacion: si topiclabel tiene NaN, esto evita que queden filas fuera del groupby
#   (si prefieres por topiclabel, deja la línea de abajo comentada y activa la alternativa)
perfil["direccionComparacion"] = (
    perfil.groupby("userid", sort=False)["direccionComparacion"].ffill().bfill()
)

# Alternativa por topic (requiere topiclabel estable):
# perfil["topiclabel"] = perfil.groupby("userid", sort=False)["topiclabel"].ffill().bfill()
# perfil["direccionComparacion"] = (
#     perfil.groupby(["userid", "topiclabel"], sort=False)["direccionComparacion"].ffill().bfill()
# )

# --- Propagar pollValue, progressMe y progressGroup con ffill() ---
perfil["pollValue_raw"] = perfil["ext_ext_responses_response"].where(perfil.verbName == "pollResponse")
perfil["progressMe_raw"] = perfil["ext_progressMe"].where(perfil.verbName == "selectSubtopic")
perfil["progressGroup_raw"] = perfil["ext_progressGroup"].where(perfil.verbName == "selectSubtopic")

perfil[["pollValue", "progressMe", "progressGroup"]] = (
    perfil.groupby(["userid", "topiclabel"], sort=False)[
        ["pollValue_raw", "progressMe_raw", "progressGroup_raw"]
    ]
    .ffill()
    .rename(
        columns={
            "pollValue_raw": "pollValue",
            "progressMe_raw": "progressMe",
            "progressGroup_raw": "progressGroup",
        }
    )
)

perfil.drop(columns=["pollValue_raw", "progressMe_raw", "progressGroup_raw"], inplace=True)

# --- Propagar pollCount con cumsum() + ffill() ---
perfil["pollCount"] = (perfil.verbName == "pollResponse").astype(int)
perfil["pollCount"] = (
    perfil.groupby(["userid", "topiclabel"], sort=False)["pollCount"]
    .cumsum()
    .ffill()
    .fillna(0)
    .astype(int)
)

# --- Helpers robustos para attempts/hints ---
def _safe_number(x, default=0):
    try:
        if pd.isna(x):
            return default
        return float(x)
    except Exception:
        return default

def _from_extra(row, key):
    try:
        raw = row.get("extra", None)
        if raw is None or (isinstance(raw, float) and pd.isna(raw)):
            return None
        data = json.loads(raw) if isinstance(raw, str) else raw
        if isinstance(data, dict):
            return data.get(key, None)
        return None
    except Exception:
        return None

def get_attempts(row):
    v = row.get("ext_attempts", np.nan)
    if pd.isna(v):
        v = row.get("ext_attemps", np.nan)  # typo tratado
    if pd.isna(v):
        v = _from_extra(row, "attempts")
    return int(_safe_number(v, 0))

def get_hints(row):
    v = row.get("ext_hints", np.nan)
    if pd.isna(v):
        v = _from_extra(row, "hints")
    return int(_safe_number(v, 0))

# --- pasos reales del ejercicio desde extra["steps"] ---
def contar_ramas_ejercicio(row):
    """
    Devuelve el número de pasos reales del ejercicio a partir de la columna 'extra',
    cuando tiene la forma:
      {"steps": {"0": {...}, "1": {...}, ...}}
    """
    raw = row.get("extra", None)
    if raw is None or (isinstance(raw, float) and pd.isna(raw)):
        return np.nan

    try:
        data = json.loads(raw) if isinstance(raw, str) else raw
    except Exception:
        return np.nan

    if isinstance(data, dict) and "steps" in data and isinstance(data["steps"], dict):
        return len(data["steps"])

    return np.nan

# --- Función para asignar sesiones ---
def asignar_sesiones(df):
    df = df.sort_values("timestamp").copy()
    df["gap_hours"] = df["timestamp"].diff().dt.total_seconds().div(3600).fillna(0)
    df["is_new_session"] = (
        (df["verbName"] == "OpenTemplateApplication") |
        (df["gap_hours"] >= 3)
    ).astype(int)
    df["session"] = df["is_new_session"].cumsum().where(lambda x: x != 0, 0)
    return df.drop(columns=["gap_hours", "is_new_session"])

perfil = (
    perfil.groupby("userid", group_keys=False, sort=False)
    .apply(asignar_sesiones)
    .reset_index(drop=True)
)

# --- Función para extraer bloques ---
def extract_blocks_usuario(df):
    bloques = []
    topic_nav = subtopic_nav = 0
    last_shownMsg = last_progressMe = last_progressGroup = None
    last_selectionRating = None
    last_ext_ext_selectionData_optionTitle = None
    poll_history_topic = {}

    for _, row in df.iterrows():
        v = row.verbName
        uid = row.userid
        topic = row.topiclabel
        key = (uid, topic)

        if key not in poll_history_topic:
            poll_history_topic[key] = []

        if v == "pollResponse" and pd.notna(row.ext_ext_responses_response):
            poll_history_topic[key].append(row.ext_ext_responses_response)

        if v == "selectTopic":
            topic_nav += 1
        elif v == "selectSubtopic":
            subtopic_nav += 1
            last_shownMsg = row.ext_shownMsg
            last_progressMe = row.ext_progressMe
            last_progressGroup = row.ext_progressGroup
        elif v == "selectionRating":
            last_selectionRating = row.result
        elif v == "displaySelection":
            last_ext_ext_selectionData_optionTitle = row.ext_ext_selectionData_optionTitle

        if v == "loadContent":
            # --- NUEVO: id del evento que abre el bloque (loadContent) ---
            bloques.append(
                {
                    "id": row.get("id", np.nan),  # <-- NUEVO (para debug)
                    "userid": uid,
                    "topiclabel": topic,
                    "start_ts": row.timestamp,
                    "end_ts": None,
                    "tryStep": 0,
                    "requestHint": 0,
                    "pollValue": row.pollValue,
                    "pollCount": row.pollCount,
                    "session": row.session,
                    "selectTopic_count": topic_nav,
                    "selectSubtopic_count": subtopic_nav,
                    "last_shownMsg": last_shownMsg,
                    "last_progressMe": last_progressMe,
                    "last_progressGroup": last_progressGroup,
                    "selectionRating_result": last_selectionRating,
                    "displaySelection_optionTitle": last_ext_ext_selectionData_optionTitle,
                    "completed": False,
                    "trayectoria": [],
                    "poll_history_topic": poll_history_topic[key].copy(),
                    "num_steps": np.nan,
                    "steps_primera": np.nan,
                    "total_attempts": np.nan,
                    "total_hints": np.nan,
                    "grupo_intervencion": row.grupo_intervencion,
                    "direccionComparacion": row.direccionComparacion,
                    "pasos_reales_ejercicio": np.nan,  # campo inicializado
                }
            )

        elif bloques:
            estado = bloques[-1]

            if v == "tryStep":
                # Guardamos stepID para poder computar steps_primera por step (NO por evento)
                estado["trayectoria"].append(
                    {
                        "evento": "tryStep",
                        "timestamp": row.timestamp,
                        "stepID": row.get("stepID", np.nan),
                        "attempts": get_attempts(row),
                        "hints": get_hints(row),
                    }
                )
            elif v == "requestHint":
                estado["trayectoria"].append(
                    {"evento": "requestHint", "timestamp": row.timestamp}
                )

            if v == "completeContent":
                try_steps = [t for t in estado["trayectoria"] if t["evento"] == "tryStep"]
                request_hints = [t for t in estado["trayectoria"] if t["evento"] == "requestHint"]

                # num_steps (eventos tryStep) y requestHint (eventos requestHint)
                estado["num_steps"] = len(try_steps)
                estado["requestHint"] = len(request_hints)

                # total attempts/hints (máximo observado en trayectoria)
                att_list = [t.get("attempts", 0) for t in try_steps]
                hint_list = [t.get("hints", 0) for t in try_steps]
                estado["total_attempts"] = int(max(att_list)) if att_list else 0
                estado["total_hints"] = int(max(hint_list)) if hint_list else 0

                # --- FIX: steps_primera por stepID (no por evento tryStep) ---
                by_step = {}
                for t in try_steps:
                    sid = t.get("stepID", np.nan)
                    if pd.isna(sid):
                        continue
                    s = by_step.setdefault(sid, {"max_attempts": 0, "max_hints": 0})
                    s["max_attempts"] = max(s["max_attempts"], int(t.get("attempts", 0)))
                    s["max_hints"] = max(s["max_hints"], int(t.get("hints", 0)))

                estado["steps_primera"] = int(
                    sum(
                        1
                        for s in by_step.values()
                        if s["max_attempts"] == 0 and s["max_hints"] == 0
                    )
                )

                # pasos reales del ejercicio desde extra["steps"]
                total_steps_real = contar_ramas_ejercicio(row)
                estado["pasos_reales_ejercicio"] = total_steps_real

                # AJUSTE: steps_primera no puede superar los pasos reales
                if not pd.isna(total_steps_real):
                    estado["steps_primera"] = min(
                        estado["steps_primera"], int(total_steps_real)
                    )

                estado["end_ts"] = row.timestamp
                estado["completed"] = True
                estado["selectionRating_result"] = last_selectionRating

                # reset contadores/estado
                topic_nav = subtopic_nav = 0
                last_shownMsg = last_progressMe = last_progressGroup = last_selectionRating = None
                last_ext_ext_selectionData_optionTitle = None

            if v in ("loadContent", "OpenTemplateApplication") and not estado["completed"]:
                estado["end_ts"] = None
                estado["completed"] = False
                topic_nav = subtopic_nav = 0
                last_shownMsg = last_progressMe = last_progressGroup = last_selectionRating = None
                last_ext_ext_selectionData_optionTitle = None

    if bloques and bloques[-1]["end_ts"] is None:
        bloques[-1]["completed"] = False

    return pd.DataFrame(bloques)

# --- Aplicar función ---
result = (
    perfil.groupby("userid", group_keys=False, sort=False)
    .apply(extract_blocks_usuario)
    .reset_index(drop=True)
)

# --- Columnas finales ---
columnas_finales = [
    "id",  # <-- NUEVA columna para debug (id del loadContent que abrió el bloque)
    "userid",
    "session",
    "topiclabel",
    "start_ts",
    "end_ts",
    "tryStep",
    "requestHint",
    "pollValue",
    "pollCount",
    "selectTopic_count",
    "selectSubtopic_count",
    "last_shownMsg",
    "last_progressMe",
    "last_progressGroup",
    "selectionRating_result",
    "displaySelection_optionTitle",
    "completed",
    "trayectoria",
    "poll_history_topic",
    "num_steps",
    "steps_primera",
    "total_attempts",
    "total_hints",
    "grupo_intervencion",
    "direccionComparacion",
    "pasos_reales_ejercicio",  # NUEVA columna
]

dfNivelEjercicio = result[columnas_finales]
```
